```python
# rewrite the HashTable class so that it uses a generator to iterate over the items in the hash table
class Node:
  def __init__(self, val):
    self.value = val
    self.next = None

class HashTable:
  def __init__(self, buckets):
    self.array = [None] * buckets
  def insert(self, val):
    bucket = hash(val) % len(self.array)
    tmp_head = Node(val)
    tmp_head.next = self.array[bucket]
    self.array[bucket] = tmp_head
  def __iter__(self):
    for i in range(len(self.array)):
      node = self.array[i]
      while node:
        yield node.value
        node = node.next
  def forEach(self, f):
    for i in range(len(self.array)):
      node = self.array[i]
      while node:
        f(node.value)
        node = node.next

# initialize the hash table
h = HashTable(3)
for i in range(10):
  h.insert(i)

# # regular
for i in h:
  for j in h:
    print(i, j)

# # generator
# def iterate(ht):
#   for i in range(len(ht.array)):
#     node = ht.array[i]
#     while node:
#       yield node.value
#       node = node.next

# # iterator
# i = iter(h)
# while True:
#   try:
#     print(next(i))
#   except StopIteration:
#     break

# for each
# h.forEach(lambda x: print(x))
```
